```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
data = pd.read_csv("week2_data.csv")

data_df = pd.DataFrame(data)
data_df
```

```python
data_df_2 = data_df.copy(deep=True)
```

```python
data_df_2 = data_df_2.drop([150000], axis= 0)
```

```python
data_df_2
```

```python
data_df_2.dropna(subset=['Bearer Id'], inplace=True)
data_df_2.dropna(subset=['MSISDN/Number'], inplace=True)
```

```python
print(data_df_2.isnull().sum())
```

```python
#number of xDR sessions
num_xdr_session = data_df_2.groupby(['MSISDN/Number'], sort = False)['Bearer Id'].agg(['count'])
num_xdr_session.columns = ['Num_of_sessions']
num_xdr_session
```

```python
#converting the microseconds to minutes
data_df_2['Dur. (mins).1'] = data_df_2['Dur. (ms).1'] / 60000000
data_df_2

#session duration for each user
session_dur = data_df_2.groupby(['MSISDN/Number'], sort = False)['Dur. (mins).1'].agg(['sum'])
session_dur.columns = ['session duration']

session_dur
```

```python
#total download(DL) and upload(UL) data for each user
total_UL_DL = data_df_2.groupby(['MSISDN/Number'], sort = False)['Total DL (Bytes)', 'Total UL (Bytes)'].agg(['sum'])
total_UL_DL.columns = ['Total DL data', 'Total UL data']

total_UL_DL
```

```python
#creating new columns in the dataframe by adding two existing columns
data_df_2['Total Social Media (Bytes)'] = data_df_2['Social Media DL (Bytes)'] + data_df_2['Social Media UL (Bytes)']
data_df_2['Total Google (Bytes)'] = data_df_2['Google DL (Bytes)'] + data_df_2['Google UL (Bytes)']
data_df_2['Total Email (Bytes)'] = data_df_2['Email DL (Bytes)'] + data_df_2['Email UL (Bytes)']
data_df_2['Total Youtube (Bytes)'] = data_df_2['Youtube DL (Bytes)'] + data_df_2['Youtube UL (Bytes)']
data_df_2['Total Netflix (Bytes)'] = data_df_2['Netflix DL (Bytes)'] + data_df_2['Netflix UL (Bytes)']
data_df_2['Total Gaming (Bytes)'] = data_df_2['Gaming DL (Bytes)'] + data_df_2['Gaming UL (Bytes)']
data_df_2['Total Other (Bytes)'] = data_df_2['Other DL (Bytes)'] + data_df_2['Other UL (Bytes)']

data_df_2
```

```python
#total data volume (in Bytes) during this session for each application
total_data_vol = data_df_2.groupby(['MSISDN/Number'], sort = False)['Total Social Media (Bytes)', 
                                                                    'Total Google (Bytes)', 
                                                                    'Total Email (Bytes)', 
                                                                    'Total Youtube (Bytes)',
                                                                    'Total Netflix (Bytes)', 
                                                                    'Total Gaming (Bytes)'].agg(['sum'])
total_data_vol.columns = ['Total Social Media Data',
                       'Total Google Data',
                       'Total Email Data',
                       'Total Youtube Data',
                       'Total Netflix Data',
                       'Total Gaming Data'
                      ]

total_data_vol
```

```python
#number of xDR sessions
plt.figure(figsize=(8, 5))
sns.set(style="darkgrid")
    
pt1 = sns.distplot(num_xdr_session['Num_of_sessions'].sample(n=50, random_state=1))
    
for item in pt1.get_xticklabels():
    item.set_rotation(45)
        
#setting the parameters for the  x and y axis labels of the plot
pt1.set_title(f"Number of xDR sessions", size=20, weight='bold')
pt1.set_xlabel("Sessions", size=10, weight='bold')
pt1.set_ylabel("Frequency", size=10, weight='bold')

plt.savefig('No_xdr_session.png')#saving the figure
plt.show()
```

```python
#session duration
plt.figure(figsize=(8, 5))
sns.set(style="darkgrid")
    
pt2 = sns.distplot(session_dur['session duration'].sample(n=50, random_state=1), color='r')
    
for item in pt2.get_xticklabels():
    item.set_rotation(45)
        
#setting the parameters for the  x and y axis labels of the plot
pt2.set_title(f"Session Duration", size=20, weight='bold')
pt2.set_xlabel("duration (mins)", size=10, weight='bold')
pt2.set_ylabel("Frequency", size=10, weight='bold')

plt.savefig('session_dur.png')#saving the figure
plt.show()
```

```python
#TOTAL UL, DL
col_name = ['Total UL data','Total DL data']  
for col in col_name:
    plt.figure(figsize=(8, 5))
    sns.set(style="darkgrid")

    pt3 = sns.distplot(total_UL_DL[col].sample(n=50, random_state=1), color='g')

    for item in pt3.get_xticklabels():
        item.set_rotation(45)

        #setting the parameters for the  x and y axis labels of the plot
    pt3.set_title(f"{col}", size=20, weight='bold')
    pt3.set_xlabel("data", size=10, weight='bold')
    pt3.set_ylabel("Frequency", size=10, weight='bold')

    plt.savefig('f{col}.png')#saving the figure
    plt.show()
```

```python
for col in total_data_vol:
    plt.figure(figsize=(8, 5))
    sns.set(style="darkgrid")

    pt4 = sns.distplot(total_data_vol[col].sample(n=50, random_state=1), color='b')

    for item in pt4.get_xticklabels():
        item.set_rotation(45)

        #setting the parameters for the  x and y axis labels of the plot
    pt4.set_title(f"{col}", size=20, weight='bold')
    pt4.set_xlabel("total data volume (bytes)", size=10, weight='bold')
    pt4.set_ylabel("Frequency", size=10, weight='bold')

    plt.savefig('f{col}.png')#saving the figure
    plt.show()
```
